# Comprensión del problema real
---

### Objetivos del negocio
Determinar la popularidad de las noticias en línea dada la cantidad de interacciones que estas han tenido en las redes sociales (número de veces que la noticia ha sido compartida). Los valores recolectados de cada noticia representan sus características como número de enlaces, número de palabras en el título, número de palabras clave, entre otros.

### Objetivos de la míneria de datos
A partir de una muestra de noticias publicadas en línea realizar una clasificación binaria: **popular vs. impopular**. Esto se debe realizar con una decisión umbral de 1400 interacciones sociales experimentando con diferentes modelos de clasificación. Lo anterior quiere decir que por debajo de 1400 interacciones la noticia es *impopular* y por encima o igual es una noticia *popular*.

### Referencias
K. Fernandes, P. Vinagre and P. Cortez. A Proactive Intelligent Decision Support System for Predicting the Popularity of Online News. Proceedings of the 17th EPIA 2015 - Portuguese Conference on Artificial Intelligence, September, Coimbra, Portugal.
- Fuente: [UCI](https://archive.ics.uci.edu/ml/datasets/Online+News+Popularity)

#### Distribución de la variable objetivo
Realizamos la reasignación de la variable objetivo 'shares' (acciones) para convertirla en una variable cualitativa nominal, con base en el umbral de distribución visto anteriormente, en la descripción de los atributos.

```python
##
## Distribución de la variable objetivo en popular/impopular
## dada la cantidad de 'shares' (variable continua) según la regla:
##
## | Condition | Popularity |
## |-----------|------------|
## | < 1400    | impopular  |
## | >= 1400   | popular    |
##

if not 'popularity' in data.columns:
    threshold = 1400
    data["popularity"] = ""   # Crea una nueva columna objetivo con la nueva clasificación
    data["popularity"][data["shares"] < threshold] = "impopular"     # Asigna clasificación de noticia "impopular"
    data["popularity"][data["shares"] >= threshold] = "popular"      # Asigna clasificación de noticia "popular"
    data.drop(columns=["shares"], inplace=True)   # Elimina columna objetivo cuantitativa 'shares'
```

#### Análisis de correlaciones
Dados los estadísticos generados anteriormente podemos concluir que existen atributos tipo ratio e indicadores que posiblemente estén **correlacionados**, por tanto, procederemos con el análisis de todas las variables cuantitativas para identificar asociaciones. A partir de las correlaciones podrémos identificar si existen variables a descartar por relaciones de linealidad que no agreguen valor al proceso de predicción.

El análisis se realizará por grupos de variables para que tenga sentido la compación utilizando el método de **Pearson**. Los grupos identificados son:
- Atributos de palabras
- Atributos de enlaces
- Atributos de palabras clave
- Atributos de PLN

```python
##
## Eliminamos las columnas categoricas para generar las correlaciones cuantitativas
##
categorial_vars = categorical_channels + categorical_weekdays
data_corr = data.drop(columns=categorical_channels)

##
## Definimos un valor límite para el valor-p de la correlación Pearson
##
p_value = 0.05

##
## Declaramos un conjunto de variables a eliminar por el análisis de correlación
##
numeric_deletes = []
```

```python
##
## Extraemos del dataset solo las variables tipo Word
##
data_corr_words = data_corr[numeric_words]

##
## Correlaciones para las variables tipo Word
##
f, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(data_corr_words.corr(method = 'pearson'), fmt="d", linewidths=.5, ax=ax)
```

A partir del gráfico anterior podemos deducir que las variables **n_non_stop_words** y **n_non_stop_unique_tokens** se encuentran fuertemente asociadas con la columna **n_unique_tokens**, por tanto, se realizará una validación del valor-p para confirmar que sea correcto prescindir de alguna en el proceso de modelado.

```python
r, p = stats.pearsonr(data['n_unique_tokens'].values, data['n_non_stop_words'].values)
print('n_unique_tokens', 'n_non_stop_words', p)
if (p < p_value) and ('n_non_stop_words' not in numeric_deletes):
    numeric_deletes.append('n_non_stop_words')

r, p = stats.pearsonr(data['n_unique_tokens'].values, data['n_non_stop_unique_tokens'].values)
print('n_unique_tokens', 'n_non_stop_unique_tokens', p)
if (p < p_value) and ('n_non_stop_unique_tokens' not in numeric_deletes):
    numeric_deletes.append('n_non_stop_unique_tokens')
    
sns.pairplot(data_corr_words[['n_unique_tokens','n_non_stop_words','n_non_stop_unique_tokens']])
```

**Conclusión:** Dado lo anterior podemos concluir que las variables **n_non_stop_words** y **n_non_stop_unique_tokens** están fuertemente relacionadas con *n_unique_tokens* y por  tanto vamos a eliminar las variables identificadas en la etapa de preparación de datos. #### Análisis de variables categóricas

Inicialmente estudiaremos los atributos relacionados con el canal o categoria de las noticias. Verificaremos la distribución de estos **canales** en la muestra y validaremos si hay completitud en los datos, ya que los canales fueron generados como variables binarias separadas por columnas.

```python
##
## Graficamos la distribución de noticias en los diferentes canales
##
data_channel = data[categorical_channels]
data_channel_patt = "data_channel_is_"
data_channel_cols = [col for col in data.columns if data_channel_patt in col]
data_channel_vals = []
data_channel_count = 0
for index, row in data_channel.iterrows():
    flag = False
    for col in data_channel_cols:
        if row[col] == 1.0:
            data_channel_vals.append(col.replace(data_channel_patt, ""))
            flag = True
            break
    if flag == False:
        # Dado que en el archivo faltan datos agregamos una nueva columna others 
        # se asigna esta cuando no exista canal entre las columnas existentes
        data_channel_vals.append("others")
        data_channel_count = data_channel_count + 1

data_channel = pd.DataFrame(data = data_channel_vals, columns = ["channel"])
data_channel["popularity"] = data.popularity
f, ax = plt.subplots(figsize=(15, 6))
sns.countplot(x="channel", hue="popularity", data=data_channel, ax=ax)

##
## Comparamos la cantidad de noticias sumando todos los canales
## con respecto al total de registros de la muestra
##
data_channel_totals = data_channel.count()
print("Total por popularidad:", data_channel_totals[1])
print("Total por canal:", (data_channel_totals[0]-data_channel_count))
print("Total datos perdidos:", data_channel_count)
```

Puede identificarse que faltan registros, como si se hubiera perdido un canal, en total son **6134** que pueden corresponder a un canal faltante. Lo representamos en la gráfica como 'others'. En la fase de preparación de datos agregaremos esta nueva columnas como canal (data_channel_is_others).

**Conclusión:** Como podemos apreciar los canales de noticias predominantes en la muestra son 'world', 'tech' y 'entertainment'. ### Selección de las características
En este proceso realizaremos dos tareas: FS (Feature Selection) y PCA (Principal Components Analysis). Se tratará de concluir si es posible agrupar características y simplificar el conjunto de datos para la fase de modelado.

Realizamos una partición de los datos en **Entrenamiento** y **Prueba** utilizando una proporción del conjunto de datos 70% y 30%, respectivamente.

```python
##
## Definimos variables para el estándar de partición 70% - 30%
##
test_size = 0.3
random_state=44

##
## Partición de los datos para probar todos los modelos
##
X_train, X_test, y_train_true, y_test_true = train_test_split(
    X, y,                       # datos originales
    test_size=test_size,        # float/int, tamaño de la muestra de prueba
    random_state=random_state)  # semilla del generador aleatorio
```

```python
##
## Construimos un Random Forest por default y modelamos el proceso
## Utilizaremos estas predicciones para identificar agrupaciones de características
##
print(__doc__)
clf = RandomForestClassifier(n_jobs = 3)
clf.fit(X_train, y_train_true)
print(clf)
y_test_pred = clf.predict(X_test)
```

**Características Importantes (FS):** El conjunto de datos actual cuenta con 51 columnas para realizar la predicción, realizaremos una comparación entre las precisiones de predicción con un único método de clasificación. Iniciaremos con la muestra completa de datos y luego haremos una búsqueda de los componentes más importantes y realizaremos predecimos de nuevo, si encontramos que no hay degradación en la precisión, continuaremos el proceso de modelado con las características filtradas como más imporantes.
- Referencia: [ScikitLearn Feature Selection](http://scikit-learn.org/stable/modules/feature_selection.html)

```python
##
## Calculamos la importancia de características
##
importances = clf.feature_importances_
std = np.std([tree.feature_importances_ for tree in clf.estimators_], axis=0)
indices = np.argsort(importances)[::-1]

#print("Feature ranking:")
#for f in range(X.shape[1]):
#    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))

##
## Extraemos las n caracteristicas que cubran el 90% de importancia acumulada
##
df = pd.DataFrame(index=indices, data=importances[indices], columns=["importance"])
df.sort_values("importance", ascending=False)
cumulative = 0
indices_new = []
for index, row in df.iterrows():
    if cumulative < 0.90:
        indices_new.append(index)
    else:
        break
    cumulative = cumulative + row['importance']
indices_new = sorted(indices_new)

##
## Entrenamos el modelo utilizando solo las columnas seleccionadas como importantes 
##
X_train_new = X_train[X_train.columns[indices_new]]
X_test_new = X_test[X_test.columns[indices_new]]
clf_new = RandomForestClassifier(n_jobs = 3)
clf_new.fit(X_train_new, y_train_true)
y_test_pred_new = clf_new.predict(X_test_new)

##
## Realizamos la predicción y comparamos la precisión del modelo para decidir
## si continuamos con las características seleccionadas como importantes o no
##
accuracy = accuracy_score(y_true = y_test_true, y_pred = y_test_pred)
accuracy_new = accuracy_score(y_true = y_test_true, y_pred = y_test_pred_new)
accuracy_diff = accuracy - accuracy_new

data_result = pd.DataFrame(data=[[accuracy, accuracy_new, accuracy_diff]],
                           columns=['Precisión', 'Presición FS', 'Diferencia'])

#print("Accuracy complete:", accuracy)
#print("Accuracy feature selection:", accuracy_new)
#print("Accuracy difference:", accuracy_diff)
#print()
flag_components = True   # Si características importantes no es viable, se realiza validación de componentes
if accuracy_diff <= -0.025:   # La diferencia debe ser más de 2.5% de la precisión con todos los datos
    X_train = X_train_new
    X_test = X_test_new
    print("El ajuste por características importantes es viable.")
    flag_components = False
else:
    print("El ajuste por características importantes NO es viable.")
    print("La diferencia debe ser más de 2.5% de la precisión con todos los datos")
    flag_components = True

data_result
```

**Conclusión:** Realizar un ajuste por *Feature Selection* no es conveniente ya que degrada la precisión de la predicción en aproximadamente **0.67%**, por tanto, se continuará el proceso de modelado sin agrupar por esta técnica.

**Componentes Principales (PCA):** Procederemos a encontrar una composición de atributos seleccionando los *k* mejores con el estadístico *Chi Square*. Si identificamos que no se degrada la predicción, procederemos a utilizar la combinación de caracterísiticas sugeridas.
- Referencia: [ScikitLearn Principal Components Analysis](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html#examples-using-sklearn-decomposition-pca)

```python
##
## Seleccionamos los k mejores componentes del conjunto de datos
##

# Utilizamos las copias para identificar si continuamos con PCA o con train originales
#X_train_org = X_train.copy()
#X_test_org = X_test.copy()

if flag_components == True:
    k_goodness = 0
    accuracy_good = 0
    rows, cols = X.shape
    list_result = []
    ## Iteramos sobre varios k tratando de encontrar el mejor ajuste
    for k in range(int(cols/2),cols,1):
        X_new = SelectKBest(chi2, k=k).fit_transform(X, y)
        X_train_new, X_test_new, y_train_true, y_test_true = train_test_split(X_new, y, 
                                                                              test_size=test_size, 
                                                                              random_state=random_state)  

        clf_new = RandomForestClassifier(n_jobs = 3)
        clf_new.fit(X_train_new, y_train_true)
        y_test_pred_new = clf_new.predict(X_test_new)

        y_test_pred_new = clf_new.predict(X_test_new)

        accuracy = accuracy_score(y_true = y_test_true, y_pred = y_test_pred)
        accuracy_new = accuracy_score(y_true = y_test_true, y_pred = y_test_pred_new)
        accuracy_diff = accuracy - accuracy_new
        
        list_result.append([k, accuracy, accuracy_new, accuracy_diff])
        #print("K-value:", k)
        #print("Accuracy complete:", accuracy)
        #print("Accuracy PCA:", accuracy_new)
        #print("Accuracy difference:", accuracy_diff)
        #print()
        if accuracy_diff <= -0.025:   # La diferencia debe ser más de 2.5% de la precisión original
            #print("El ajuste por combinación de componentes es viable.")
            if accuracy_good < abs(accuracy_diff):
                X_train = X_train_new
                X_test = X_test_new
                k_goodness = k
    if k_goodness > 0:
        print("El ajuste por combinación de componentes es viable.")
    else:
        print("El ajuste por combinación de componentes NO es viable.")
        print("La diferencia debe ser más de 2.5% de la precisión con todos los datos")        
    print("El mejor valor de k encontrado es:", k_goodness)
else:
    print("No se realiza combinación de componentes, dado que características importantes es viable")
    
data_result = pd.DataFrame(data=list_result,
                           columns=['Valor K', 'Precisión', 'Presición FS', 'Diferencia'])
data_result
```

**Conclusión:** Los calculos anteriores permiten realizar las predicciones en la fase de modelado utilizando una combinación de características, sin embargo, el resultado anterior indica que **no existe una combinación de componentes** que asegure un mejor ajuste en las predicciones, por lo tanto, se continua con todo el conjunto de atributos original.

```python
##
## Construimos una función personalizada para ejecutar la metodología
##
def grid_search_custom(classifier,
                       tuned_params,
                       scores):
    #start = tm.time() % 60
    scores_result = []
    for score in scores:
        print()
        print("Parámetros ajustados del score: %s" % score)
        print()

        # Entrenamiento utilizando cross-validation (folds = 3)
        clf = GridSearchCV(classifier, tuned_params, cv=3, scoring='%s_macro' % score, n_jobs=3)
        clf.fit(X_train, y_train_true)
        
        print("Mejores parámetros encontrados:")
        print(clf.best_params_)
        print()
        
        # Predicción con datos de prueba para validar metricas
        y_true, y_pred = y_test_true, clf.predict(X_test)
        
        #print("Reporte de clasificación detallado:")
        #print()
        #print(classification_report(y_true, y_pred))
        #print()
        
        accuracy_result = accuracy_score(y_true, y_pred)
        recall_result = recall_score(y_true, y_pred)
        f1_result = f1_score(y_true, y_pred)
        roc_auc_result = roc_auc_score(y_true, y_pred)
        jaccard_result = jaccard_similarity_score(y_true, y_pred)
        conf_matrix_result = confusion_matrix(y_true, y_pred)
        
        scores_result.append([score, 
                              round(accuracy_result, 2), 
                              round(1-accuracy_result, 2), 
                              round(recall_result, 2), 
                              round(f1_result, 2), 
                              round(roc_auc_result, 2), 
                              round(jaccard_result, 2),
                              clf.best_params_,
                              conf_matrix_result])
    
    return(scores_result)
    #end = tm.time() % 60
    #print("Tiempo total de ejecución (segundos): %.2f" % (end - start))
```

          N  |  TN    FP

- FN: Falso negativo (mal clasificado)
- FP: Falso positivo (mal clasificado)

```python
##
## Función personalizada para calcular el CrossTable de R
## Solo arma la tabla para un objetivo binario
##
def cross_table_custom(cnf_matrix,
                       neg_label,
                       pos_label):
    tn, fp, fn, tp = cnf_matrix.ravel()
    names = [[neg_label for i in range(4)], [pos_label for j in range(4)], ["ColumnsTotal" for k in range(2)]]
    index1 = [val for sublist in names for val in sublist]
    index2 = [" " for idx in range(len(index1))]
    index = pd.MultiIndex.from_arrays([index1, index2])
    columns = [neg_label, pos_label, "RowsTotal"]
    
    str_ints = "{:.0f}"
    str_float = "{:.2f}"
    sum_all = tp+fp+fn+tn
    sum_tnfp = tn+fp
    sum_tnfn = tn+fn
    sum_tpfp = tp+fp
    sum_tpfn = tp+fn
    
    values = [[str_ints.format(tn),str_ints.format(fp),str_ints.format(sum_tnfp)],
              [str_float.format(tn/sum_tnfp),str_float.format(fp/sum_tnfp),str_float.format(sum_tnfp/sum_all)],
              [str_float.format(tn/sum_tnfn),str_float.format(fp/sum_tpfp)," "],
              [str_float.format(tn/sum_all),str_float.format(fp/sum_all)," "],
              [str_ints.format(fn),str_ints.format(tp),str_ints.format(sum_tpfn)],
              [str_float.format(fn/sum_tpfn),str_float.format(tp/sum_tpfn),str_float.format(sum_tpfn/sum_all)],
              [str_float.format(fn/sum_tnfn),str_float.format(tp/sum_tpfp)," "],
              [str_float.format(fn/sum_all),str_float.format(tp/sum_all)," "],
              [str_ints.format(sum_tnfn),str_ints.format(sum_tpfp),str_ints.format(sum_all)],
              [str_float.format(sum_tnfn/sum_all),str_float.format(sum_tpfp/sum_all)," "]]
    return pd.DataFrame(index=index, columns=columns, data=values)

cross_table_custom(conf_matrix[2], "Impopular", "Popular")
```
